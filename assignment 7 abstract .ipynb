```python
'''Q3. What is abc module in python? Why is it used?'''

#Answer:

'''In Python, the abc module stands for "Abstract Base Classes." It provides a way to define abstract classes 
in Python, which are classes that cannot be instantiated directly but serve as a blueprint for subclasses. 
The primary purpose of the abc module is to enable the creation of interfaces and enforce the implementation 
of certain methods in subclasses.

Here are some key points about the abc module and its usage:

Abstract Base Classes (ABCs): An abstract base class is defined using the ABC metaclass provided by the abc module.
It allows you to specify methods that must be implemented by any concrete subclass. ABCs define a common API and 
establish a contract for subclasses.

Abstract Methods: Abstract methods are declared using the @abstractmethod decorator, which marks them as methods 
that must be implemented by any concrete subclass. These methods raise a TypeError if they are not overridden by 
the subclass.  '''
```

```python
''' Q4. How can we achieve data abstraction?'''


#Answer : 



'''In Python, data abstraction can be achieved through the use of classes and objects. Abstraction is a fundamental 
principle of object-oriented programming (OOP) that allows you to represent complex systems by hiding unnecessary details 
and exposing only the essential features.

Here are some key concepts and techniques for achieving data abstraction in Python:

Classes and Objects: Classes are the blueprints that define the structure and behavior of objects, while objects are instances 
of classes. Classes encapsulate data (attributes) and behavior (methods) into a single entity. By creating classes, you can 
abstract away the implementation details and provide a clean interface for working with objects.

Encapsulation: Encapsulation is the process of bundling data and related methods together within a class, thereby hiding the 
internal details of the class from outside access. It allows you to control the access to data and enforce proper usage through 
methods. Encapsulation helps in achieving information hiding and protecting the integrity of the data.

Access Modifiers: In Python, access modifiers are achieved through naming conventions rather than language-enforced keywords like 
in some other programming languages. By convention, attributes and methods that start with a single underscore _ are considered 
"internal" and should not be accessed directly from outside the class. Attributes and methods starting with double underscores
 __ (name mangling) are considered "private" and have limited access. This convention helps in controlling access to data and 
 promoting encapsulation.

Getter and Setter Methods: Getter and setter methods provide controlled access to class attributes. They allow you to retrieve 
the value of an attribute (getter) or modify its value (setter) while enforcing any necessary validations or transformations. 
By using getter and setter methods, you can abstract away the details of how the attribute is stored and provide a consistent 
interface for accessing and modifying data.'''
```

```python
# Q5. Can we create an instance of an abstract class? Explain your answer.


#answer 

''' No, you cannot create an instance of an abstract class in Python. Abstract classes are designed to serve as blueprints 
for subclasses and cannot be instantiated directly. They are meant to be inherited by concrete subclasses that provide 
implementations for the abstract methods defined in the abstract class.

Here are a few reasons why you cannot create an instance of an abstract class:

Incomplete Implementation: Abstract classes contain one or more abstract methods, which are methods declared without an 
implementation. These methods act as placeholders and must be implemented by concrete subclasses. Since abstract classes 
have incomplete implementations, they cannot be instantiated.

Interface Definition: Abstract classes often define interfaces or contracts that specify a set of methods that must be 
implemented by subclasses. Instantiating an abstract class would mean creating an object that doesn't fully adhere to the 
interface contract, as it would lack the implementation of one or more abstract methods. '''
```
